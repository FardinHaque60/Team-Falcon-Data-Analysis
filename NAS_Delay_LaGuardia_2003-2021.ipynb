# NAS Delay in LaGuardia International Airport Data Analysis

```python
import csv 

with open('LaGuardia_Data.csv') as csvfile:
    LaGuardia_Data = list(csv.DictReader(csvfile))
```

## Iterating through data to create dictionary:

```python
temp_list1 = []
temp_list2 = []
for n in range(0, len(LaGuardia_Data)):
    temp_dict = LaGuardia_Data[n]
    temp_list1.append(int(temp_dict["year"]))
    try:
        temp_list2.append(float(temp_dict["nas_delay"]))
    except ValueError:
        temp_list2.append(None)
```

## Putting dictionary in pandas datatable:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dict_data = {"year": temp_list1,
            "nas_delay": temp_list2}

usable_data = pd.DataFrame(data=dict_data)
usable_data
```

## Plotting data using matplotlibs:

```python
x = usable_data.year
y = usable_data.nas_delay
```

```python
plt.scatter(x,y, marker="o", alpha=0.4, s=20)
plt.xlabel("Y e a r")
plt.ylabel("D e l a y (min)")
plt.ylim(0,140000)
plt.xticks(np.arange(min(x), max(x)+1, 2.0))
```

```python
plt.subplot(211)
plt.scatter(x,y, marker="o", alpha=0.4, s=20)
plt.ylim(10000,50000)
plt.xlim(2015,)
plt.ylabel("D e l a y (min)")
plt.xticks(np.arange(2012, 2020, 1.0))
#top half (recent years)

plt.subplot(212)
plt.scatter(x,y, marker="o", alpha=0.4, s=20)
plt.ylim(0,10000)
plt.xlim(2015,)
plt.xlabel("Y e a r")
plt.ylabel("D e l a y (min)")
plt.xticks(np.arange(2012, 2020, 1.0))
#bottom half (recent years)
```
